Carga del dataset

```python
import pandas as pd

data = pd.read_csv("Dataset/movies_dataset.csv")
data.head(3)
```

Elimino las filas que tienen los datos en orden diferente al indicado en el header.

```python
data = data[data["adult"].isin(["True", "False"])]
data.info()
```

Eliminar las columnas que no serán utilizadas, video, imdb_id, adult, original_title, poster_path y homepage.

```python
data.drop(columns=["video", "imdb_id", "adult", "original_title", "poster_path", "homepage"], inplace=True)
data.info()
```

Los valores nulos del campo release date deben eliminarse.

```python
data.dropna(subset="release_date", inplace=True)
data.info()
```

Los valores nulos de los campos revenue, budget deben ser rellenados por el número 0.

```python
data["revenue"] = data["revenue"].fillna(0)
data["budget"] = data["budget"].fillna(0)
```

Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, dividiendo estas dos últimas revenue / budget, cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0.

```python
data["budget"] = data["budget"].apply(pd.to_numeric)
data["return"] = [x / y if y > 0 else 0 for x, y in zip(data["revenue"], data["budget"])]
data.info()
```

De haber fechas, deberán tener el formato AAAA-mm-dd, además deberán crear la columna release_year donde extraerán el año de la fecha de estreno.

```python
data["release_date"] = pd.to_datetime(data["release_date"])
data.info()
```

```python
import datetime as dt
data["release_year"] = data["release_date"].dt.strftime('%Y')
data.head(3)
```

```python
import ast
import numpy as np

data['belongs_to_collection'] = data['belongs_to_collection'].apply(lambda x: ast.literal_eval(x)['name'] if pd.notnull(x) else np.nan)
data.head(3)
```

```python
data["production_countries"] = data["production_countries"].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else np.nan)
data["production_companies"] = data["production_companies"].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else np.nan)
data.head(3)
```

```python
datac = pd.read_csv("Dataset/credits.csv")
datac["id"] = datac["id"].apply(pd.to_numeric)
data["id"] = data["id"].apply(pd.to_numeric)
datac.info()
```

```python
data_all = pd.merge(data, datac, on="id", how="left")
data_all["crew"] = data_all["crew"].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else np.nan)
data_all.info()
```

```python
data_all.head()
```

```python
data_all.drop(data_all[data_all["crew"].isnull()].index, inplace=True)
data_all.drop(data_all[data_all['crew'].apply(len) == 0].index, inplace=True)
# Creamos una columna nueva con un indicador de si se participó como director en dicha película.
data_all["is_director_in"] = data_all["crew"].apply(lambda x: np.count_nonzero([1 for n in x if (n["job"] == "Director") and (n["name"] == "Christopher Nolan")]))
# Calculamos cantidad de películas en las que participó y retorno general.
movie_qty = data_all["is_director_in"].sum()
total_return = data_all.loc[data_all["is_director_in"] == 1]["revenue"].sum() / data_all.loc[data_all["is_director_in"] == 1]["budget"].sum()
# Obtenemos la lista de películas con la fecha de lanzamiento, retorno individual, costo y ganancia.
data_draft = data_all.loc[data_all["is_director_in"] == 1]
data_draft = data_draft.drop(columns=['belongs_to_collection','genres', 'id', 'original_language', 'overview', 'popularity', 'production_companies',
 'production_countries', 'runtime', 'spoken_languages', 'status', 'tagline', 'vote_average', 'vote_count', 'release_year', 'cast', 'crew', 'is_director_in'])
data_draft
```

```python
new_list = [{"title": data["title"], "release_date": pd.to_datetime(data["release_date"]).strftime("%Y/%m/%d"), "budget": data["budget"], "revenue": data["revenue"], "return": round(data["return"],2)} for (index, data) in data_draft.iterrows()]
new_list
```
